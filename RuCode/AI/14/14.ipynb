```python
import pandas as pd
import numpy as np
train = pd.read_csv("train_data.csv").dropna(how='any')

X = train.drop(columns=['Цена продажи', 'Качество'])
y = train['Цена продажи']
# qual = train['Качество']

X_test = pd.read_csv("test_data.csv").dropna(how='any').drop(columns=['Качество'])
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
gr = LinReg.fit(X, y)
pred = LinReg.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
cor = train.corr()
plt.imshow(cor, cmap='RdBu', interpolation='nearest')
plt.colorbar()
```

```python
fig, ax = plt.subplots()
ax.bar(range(len(pred)), pred, label='MLP_prediction')
fig.set_figwidth(20)
fig.set_figheight(20)
plt.legend()
```

```python
X_test['Спальни'] = pred
```
